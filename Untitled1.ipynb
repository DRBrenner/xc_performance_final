```python
import numpy as np
import pandas as pd
```

```python
log2 = pd.read_csv('db/training_log_part2.csv')
log2
```

```python
log1 = pd.read_csv('db/training_log_part1.csv')
log1
```

```python
log = pd.concat([log1,log2])
log
```

```python
# Add month and year from meet_date
def get_month(x):
    return int(str(x).split('/')[0])
log['month'] = log['log_date'].apply(get_month)

def get_year(x):
    return int(str(x).split('/')[2])
log['year'] = log['log_date'].apply(get_year)

log.tail()
```

```python
log_2015_to_2018 = log.loc[(log['year'] > 2014) & (log['year'] < 2019)]
log_2015_to_2018
```

```python
log_2015_to_2018.to_csv('db/log_2015_to_2018.csv')
```
