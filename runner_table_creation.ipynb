```python
results_df = pd.read_csv('db/results_2015_to_2018.csv')
results_df.head()
```

```python
# Delete unused columns
results_df = results_df.drop(results_df.columns[0], axis=1)
del results_df['previous_time']
del results_df['last_year_time']
del results_df['improvement_flag']
results_df = results_df.rename(columns = {"pg_flag":"goal_flag"})
results_df.head()
```

```python
# Change Null to "0" in goal columns
results_df["goal_time"] = results_df['goal_time'].replace(np.nan, 0)
results_df["goal_flag"] = results_df['goal_flag'].replace(np.nan, 0)
results_df.head()
```

```python
# Drop rows with race_result = 0
results_df = results_df[results_df.race_result != 0].reset_index()
```

```python
# Change gender to all lowercase
results_df["gender"] = results_df['gender'].str.lower()
results_df.gender[results_df.gender == 'b'] = 1
results_df.gender[results_df.gender == 'g'] = 0
results_df.head()
```

```python
results_df['goal_set'] = np.where(results_df['goal_time'] > 0, 1, 0)
```

```python
results_df["pace"] = (results_df["race_result"])/((results_df["distance"])*60)
results_df.head()
```

```python
# Copy meet_date and convert to UNIX format so I can filter by date.
results_df["datetime"] = pd.to_datetime(results_df['meet_date']) #required to use < or > on date
results_df = results_df.sort_values(['datetime','course_id','distance', 'gender','race_result'])
results_df.head()
```

```python
grouped =  results_df.groupby(['datetime', 'course_id',"distance","gender"])
results_df['Place'] = grouped['race_result'].transform(lambda x: pd.factorize(x, sort=True)[0]+1)
results_df['Rank'] = grouped['race_result'].rank(method='min')
results_df['pct_rank'] = grouped['race_result'].rank(method='min',pct=True)
results_df.sort_values(['datetime', 'course_id',"distance","gender","Rank"])
```

```python
all_300plus_races = results_df.loc[(results_df["distance"]>=300)]
all_races_winners = all_300plus_races.groupby(['datetime', 'course_id',"distance","gender"]).first().reset_index()
all_races_winners
```

```python
all_races_winners_average = all_races_winners.groupby(["year","gender"])["pace"].mean().reset_index()
all_races_winners_average
```

```python
# Merge total runner count (per school) via school_id to results_df
results_complete_df = pd.merge(results_df, school_runner_count_fall_2018, how="left", on="school_id")
results_complete_df = pd.merge(results_df, school_runner_count_all, how="left", on="school_id")

# Merge log entry data (per runner) via runr_id to results_complete_df - identical log entries per runner's race result
results_complete_df = pd.merge(results_complete_df, log_count_9, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, log_count_11, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, log_count_all, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, max_effort_log_count_9, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, max_effort_log_count_11, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, max_effort_log_count_all, how="left", on="runr_id")
results_complete_df = pd.merge(results_complete_df, log_sum_11, how="left", on="runr_id")

# # Change Null to "0" in log columns
results_complete_df["log_entries_sept"] = results_complete_df['log_entries_sept'].replace(np.nan, 0)
results_complete_df["log_entries_nov"] = results_complete_df['log_entries_nov'].replace(np.nan, 0)
results_complete_df["log_entries_all"] = results_complete_df['log_entries_all'].replace(np.nan, 0)
results_complete_df["max_effort_log_entries_sept"] = results_complete_df['max_effort_log_entries_sept'].replace(np.nan, 0)
results_complete_df["max_effort_log_entries_nov"] = results_complete_df['max_effort_log_entries_nov'].replace(np.nan, 0)
results_complete_df["max_effort_log_entries_all"] = results_complete_df['max_effort_log_entries_all'].replace(np.nan, 0)
results_complete_df["sum_effort_log_entries_nov"] = results_complete_df['sum_effort_log_entries_nov'].replace(np.nan, 0)

results_complete_df

```

```python
# Create "race_diff" field to calculate the different between best time (pr_time) and current race time for that course
results_complete_df["race_diff"] = results_complete_df["pr_time"] - results_complete_df["race_result"]
results_complete_df.tail()
```

```python
# Get all 2015-2018 races and calculate the total number of races per year per racer ("count") to calculate rounded one third of races
runner_race_counts = results_complete_df.groupby(["runr_id","year","datetime","course_id","pace"]).nunique()["race_diff"].reset_index()
# runner_race_counts2018 = runner_race_counts.loc[(runner_race_counts['datetime']>'2018-5-31') & (runner_race_counts['datetime']<'2019-1-1')]
runner_race_counts["count"] = runner_race_counts.groupby(["runr_id","year"])['runr_id'].transform('count')
runner_race_counts["one_third_count"] = np.ceil(runner_race_counts["count"]/3)

```

```python
# THIS CODE WORKS, BUT CANNOT FIGURE OUT HOW TO CALL THE COLUMNS (even to rename), SO NOT USING
runner_race_counts_end_year = runner_race_counts.copy()

#Group by runr_id and year, but keep only last third of records.  The calculate the mean of those records (per year per runr_id)
runner_race_counts_end_year= runner_race_counts_end_year.groupby(["runr_id","year"]).tail(runner_race_counts_end_year["one_third_count"])
runner_race_counts_end_year= runner_race_counts_end_year.groupby(["runr_id","year"])["pace"].mean().reset_index().sort_values("runr_id")

#Reshape (Pivot) year rows to be columns
runner_race_counts_end_year = runner_race_counts_end_year.pivot(index="runr_id",columns="year",values="pace").reset_index()
runner_race_counts_end_year = runner_race_counts_end_year.rename(columns = {"2015":"end_2015"})
runner_race_counts_end_year.tail()
```

```python
# Get all 2018 races and calculate the total number of races per racer ("count") to calculate rounded one third of races
runner_race_counts = results_complete_df.groupby(["runr_id","datetime","course_id","pace"]).nunique()["race_diff"].reset_index()
runner_race_counts2018 = runner_race_counts.loc[(runner_race_counts['datetime']>'2018-5-31') & (runner_race_counts['datetime']<'2019-1-1')]
runner_race_counts2018["count"] = runner_race_counts2018.groupby("runr_id")['runr_id'].transform('count')
runner_race_counts2018["one_third_count"] = np.ceil(runner_race_counts2018["count"]/3)
runner_race_counts2018.tail()
```

```python
# Calculate average pace of last third of races per racer ("2018_end_pace")
runner_race_counts_end = runner_race_counts2018.copy()

runner_race_counts_end["2018_end_pace"] = runner_race_counts_end.groupby("runr_id")['pace'].tail(runner_race_counts_end["one_third_count"])
runner_race_counts_end = runner_race_counts_end.dropna(subset=['2018_end_pace'])
runner_race_counts_end = runner_race_counts_end.groupby("runr_id")['2018_end_pace'].mean().reset_index()
runner_race_counts_end.tail()
```

```python
# Calculate average pace of first third of races per racer ("2018_beg_pace")
runner_race_counts_beg = runner_race_counts2018.copy()

runner_race_counts_beg["2018_beg_pace"] = runner_race_counts_beg.groupby("runr_id")['pace'].head(runner_race_counts_beg["one_third_count"])
runner_race_counts_beg = runner_race_counts_beg.dropna(subset=['2018_beg_pace'])
runner_race_counts_beg = runner_race_counts_beg.groupby("runr_id")['2018_beg_pace'].mean().reset_index()
runner_race_counts_beg.tail()
```

```python
# Create new runners table with aggregated information.  Clean up table
runners_table_2018 = results_complete_df.loc[(results_complete_df['datetime']>'2018-5-31') & (results_complete_df['datetime']<'2019-1-1')]

aggregator = {'school_id':'max',
              'runner_count':'max',
              'gender':'max',
              'pace' : 'mean',
              'datetime': 'count',
              'Rank':'mean',
              'pct_rank':'mean',
              'pr_time':'min',
              'race_diff':'mean',
              'pr_flag':'sum',
              'goal_set':'sum',
              'log_entries_sept':'max',
              'log_entries_nov':'max',
              'log_entries_all':'max',
              'max_effort_log_entries_sept':'max',
              'max_effort_log_entries_nov':'max',
              'max_effort_log_entries_all':'max',
              'sum_effort_log_entries_nov':'max'
             }
runners_2018 = runners_table_2018.groupby(["runr_id"])['school_id','runner_count','gender',"pace",'datetime','Rank','pct_rank','pr_time',
                                                           'race_diff','pr_flag','goal_set','log_entries_sept','log_entries_nov','log_entries_all',
                                                          'max_effort_log_entries_sept','max_effort_log_entries_nov',
                                                          'max_effort_log_entries_all','sum_effort_log_entries_nov'].agg(aggregator).reset_index()

runners_2018 = runners_2018.rename(columns = {"runner_count":"team_size","datetime":"num_races","Rank":"rank"})
runners_2018['pace'] = runners_2018['pace'].round(2)
runners_2018['rank'] = runners_2018['rank'].round(1)
runners_2018['pct_rank'] = runners_2018['pct_rank'].round(2)
runners_2018['race_diff'] = runners_2018['race_diff'].round(2)
runners_2018.tail()
```

```python
# Merge runner tables
runners_2018_complete = pd.merge(pd.merge(runner_race_counts_beg,runner_race_counts_end,on='runr_id'),runners_2018,on='runr_id')
runners_2018_complete["2018_improvement"] = runners_2018_complete["2018_beg_pace"] - runners_2018_complete["2018_end_pace"]
runners_2018_complete["2018_improvement"] = runners_2018_complete["2018_improvement"].round(2)
runners_2018_complete.tail()
```
