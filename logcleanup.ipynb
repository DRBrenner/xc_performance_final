```python
import numpy as np
import pandas as pd
```

```python
log2 = pd.read_csv('db/training_log_part2.csv')
log2.head()
```

```python
log1 = pd.read_csv('db/training_log_part1.csv')
log1.head()
```

```python
# Append log2 to log1
log = pd.concat([log1,log2])
log.head()
```

```python
# Add month and year from log_date
def get_month(x):
    return int(str(x).split('/')[0])
log['month'] = log['log_date'].apply(get_month)

def get_year(x):
    return int(str(x).split('/')[2])
log['year'] = log['log_date'].apply(get_year)

log.tail()
```

```python
log_2015_to_2018 = log.loc[(log['year'] > 2014) & (log['year'] < 2019)]
log_2015_to_2018.head()
```

```python
results_all = pd.read_csv('db/results_and_goals.csv')
results_all.head()
```

```python
# Add month and year from meet_date
def get_year(x):
    return int(str(x).split('/')[2])
results_all['year'] = results_all['meet_date'].apply(get_year)

def get_month(x):
    return int(str(x).split('/')[0])
results_all['month'] = results_all['meet_date'].apply(get_month)

results_all.head()
```

```python
results_df = results_all.loc[(results_all['year'] > 2014) & (results_all['year'] < 2019)].sort_values(["year","month"])
results_df
```

```python
# Write new log csv to db folder
log_2015_to_2018.to_csv('db/log_2015_to_2018.csv')
```

```python
# Write new results csv to db folder
results_df.to_csv('db/results_2015_to_2018.csv')
```
